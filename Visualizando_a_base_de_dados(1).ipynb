```python
btc_latan_volume = 0
eth_latan_volume = 0
ltc_latan_volume = 0

for i in amount_btc.sum():
    btc_latan_volume += i

for i in amount_eth.sum():
    eth_latan_volume += i

for i in amount_ltc.sum():
    ltc_latan_volume += i
 
```

```python
volume_btc = []
volume_eth = []
volume_ltc = []
for i in amount_btc:
    per_volume_btc = amount_btc[i].sum()/btc_latan_volume
    volume_btc.append(per_volume_btc)
    
for i in amount_eth:
    per_volume_eth = amount_eth[i].sum()/eth_latan_volume
    volume_eth.append(per_volume_eth)

for i in amount_ltc:
    per_volume_ltc = amount_ltc[i].sum()/ltc_latan_volume
    volume_ltc.append(per_volume_ltc)
```

```python
porcetagem_mercado_btc= {"Bitcoin Trade":None,
                        "Bitso":None,
                        "Brazillex":None,
                        "CryptoMarket":None,
                        "Buda":None,
                        "Mercado Bitcoin":None}

porcetagem_mercado_eth= {"Bitcoin Trade":None,
                        "Bitso":None,
                        "Brazillex":None,
                        "CryptoMarket":None,
                        "Buda":None,
                        "Mercado Bitcoin":None}

porcetagem_mercado_ltc= {"Bitcoin Trade":None,
                        "Bitso":None,
                        "Brazillex":None,
                        "CryptoMarket":None,
                        "Buda":None,
                        "Mercado Bitcoin":None}
```

```python
index = 0
for k in porcetagem_mercado_btc.keys():
    try:
        porcetagem_mercado_btc[k] = volume_btc[index]
        porcetagem_mercado_ltc[k] = volume_ltc[index]
        porcetagem_mercado_eth[k] = volume_eth[index]
    except IndexError:
        pass
    index+=1
```

```python
exchanges = ['bitcointrade','bitso','brazillex','cryptomarket','buda','mercado']

bitcoin = pd.DataFrame(porcetagem_mercado_btc, index=["Porcetagem"])
ethereum = pd.DataFrame(porcetagem_mercado_eth, index=['Porcetagem'])
litcoin = pd.DataFrame(porcetagem_mercado_ltc,index=['Porcetagem']).T
```

```python
fig, ax = plt.subplots(figsize=(12, 7), subplot_kw=dict(aspect="equal"), dpi= 80)

dados = bitcoin['Porcetagem']
categorias  = bitcoin.index
explode = [0,0,0,0,0,0.1]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%".format(pct)


wedges, texts, autotexts = ax.pie(dados, 
                                  autopct=lambda pct: func(pct, dados),
                                  textprops=dict(color="w"), 
                                  colors=plt.cm.Dark2.colors,
                                 startangle=140,
                                 explode=explode)


ax.legend(wedges, categorias, title="Volume Share de Bitcoin na LATAM", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=10, weight=700)
ax.set_title("Volume Share: Pie Chart")
plt.tight_layout()
```
